```python
data=pd.read_csv("result.csv")
NSIDE=32
NPIX=hp.nside2npix(NSIDE)
```

```python
# dec = np.array([])
# ra = np.array([])
```

```python
ra = data['RAJ2000'].tolist()
dec = data['DEJ2000'].tolist()
```

```python
indices = hp.ang2pix(NSIDE, ra, dec, lonlat=True)
indices
```

```python
mapplot = np.zeros(NPIX)

for i in indices:
    mapplot[i] = mapplot[i] + 1
```
